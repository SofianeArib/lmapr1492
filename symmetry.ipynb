```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from pymatgen.ext.matproj import MPRester
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
with MPRester ("puId7PZyCiPcHGVBKl") as m:
    #On définit les variables nécessaires pour la suite
    structure = m.get_structure_by_material_id("mp-9366")
    ConventionalStructure = SpacegroupAnalyzer(structure).get_conventional_standard_structure()
    analyzer = SpacegroupAnalyzer(ConventionalStructure)
    SymmetryList = analyzer.get_symmetry_operations() #SymmOp object
```

```python
#Groupe de symétrie ?
print("Le groupe de symétrie est :", analyzer.get_space_group_symbol())
print("Pour la suite, la position des atomes ainsi que les images de la maille conventionnelle sont obtenues à l'aide du site Jmol.")
```

```python
#1ère symétrie (de translation) sur l'atome de Rb situé en [0,0,0]
Translation = SymmetryList[24]
print("On étudie une symétrie de translation sur l'atome de rubidium situé en [0,0,0]:")
print('\n',Translation.rotation_matrix,"[0,0,0]","+",Translation.translation_vector,"=",Translation.operate([0,0,0]))
print("\n Sur l'image, on voit que la position finale est [1/3,-1/3,-1/3] ce qui correspond aux mêmes coordonées réduites après translations par les vecteurs de base b et c.")
plt.figure(figsize = (9,9))
plt.imshow(mpimg.imread('Rb_Translation.png'))
plt.show()
```

```python
#2e symétrie (ponctuelle) sur l'atome de S situé en [2/3,1/3,1/10]
Mirror = SymmetryList[7]
print("On étudie une symétrie ponctuelle miroir sur l'atome de soufre situé en [2/3,1/3,1/10]:")
print('\n',Mirror.rotation_matrix,"[2/3,1/3,1/10]","+",Mirror.translation_vector,"=",Mirror.operate([2/3,1/3,1/10]))
print("\n En translatant par les vecteurs de base a et b, on obtient les coordonées réduites [2/3,1/3,1/10]. On observe bien sur l'image que le point d'arrivée occupe la même position dans la maille conventionnelle que l'atome de départ.")
plt.figure(figsize = (9,9))
plt.imshow(mpimg.imread('S_Mirror.png'))
plt.show()
```

```python
#3e symétrie (spatiale) sur l'atome de Ho situé en [1/3,2/3,1/6]
dglide = SymmetryList[31]
print("On étudie une symétrie spatiale qui est la réflexion avec glissement sur l'atome de holmium situé en [1/3,2/3,1/6]:")
print('\n',dglide.rotation_matrix,"[1/3,2/3,1/6]","+",dglide.translation_vector,"=",dglide.operate([1/3,2/3,1/6]))
print("\n En translatant par le vecteur de base a, on trouve les coordonées réduites [2/3,1/3,5/6] qui est bien la position de l'atome de holmium indiqué sur l'image.")
plt.figure(figsize = (9,9))
plt.imshow(mpimg.imread('Ho_dglide.png'))
plt.show()
```
