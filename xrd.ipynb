```python
import matplotlib.pyplot as plt
from pymatgen.analysis.diffraction.xrd import *
from pymatgen.analysis.diffraction.core import *
from pymatgen.ext.matproj import MPRester
from pymatgen.symmetry.analyzer import SpacegroupAnalyzer
with MPRester ("puId7PZyCiPcHGVBKl") as m:
    #On définit les variables nécessaires pour la suite
    structure = m.get_structure_by_material_id("mp-9366")
    ConventionalStructure = SpacegroupAnalyzer(structure).get_conventional_standard_structure()
    xrd = XRDCalculator(1.5406) #XRD Calculator object
    pattern = xrd.get_pattern(ConventionalStructure) #renvoie un object DiffractionPattern
```

```python
#L'objet DiffractionPattern se trouve dans le module pymatgen.analysis.diffraction.core
theta = pattern.x #Tableau contenant les angles 2*theta
hkl = pattern.hkls #Tabeau de tableaux contenant les différents indices hkl ainsi que leur multiplicité
```

```python
#On reprend les 3 premiers pics et on les affiche
print("\n Le premier pic sera observé avec un angle 2θ =",theta[0],"° et correspond au plan",hkl[0][0]["hkl"])
print("\n Le second pic sera observé avec un angle 2θ =",theta[1],"° et correspond au plan",hkl[1][0]["hkl"])
print("\n Le troisième pic sera observé avec un angle 2θ =",theta[2],"° et correspond au plan",hkl[2][0]["hkl"])
print("\n On remarque un 4e indice dans la notation des plans. Celui-ci est utilisé pour les systèmes hexagonal et \n rhombohédrique et se note (hkil) avec i qui est un indice redondant. Il respecte l'équation h + k + i = 0. \n Cet indice sert à faire apparaitre les symétries de permutation.")
```
